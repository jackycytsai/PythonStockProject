```python
import pandas as pd
import numpy as np
import requests
from io import StringIO
import datetime
```

```python
def StockTable(date):
    #date = 20230922
    datestr = date

    # 下載股價
    r = requests.post('https://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + datestr + '&type=ALLBUT0999')

    # 整理資料，變成表格
    df = pd.read_csv(StringIO(r.text.replace("=", "")), 
                header=["證券代號" in l for l in r.text.split("\n")].index(True)-1)

    field_names = [
    'code',
    'name',
    'tradingVolume',
    'numberOfTrades',
    'tradingAmount',
    'openingPrice',
    'highestPrice',
    'lowestPrice',
    'closingPrice',
    'priceChange',
    'priceChangeDifference',
    'lastRevealedBuyingPrice',
    'lastRevealedBuyingQuantity',
    'lastRevealedSellingPrice',
    'lastRevealedSellingQuantity',
    'priceToEarningsRatio',
    'Date'
    ]

    # Rename the columns of the DataFrame
    df.columns = field_names

    # Remove Column
    df = df.drop(columns='priceChange')

    # 整理一些字串：
    df[df.columns[2:]] = df[df.columns[2:]].applymap(lambda x: pd.to_numeric(str(x).replace(",", "").replace("+", "1").replace("-", "-1"), errors='coerce'))

    # Insert Date
    df.Date = datestr

    # Convert the 'Date' column to datetime format
    df['Date'] = pd.to_datetime(df['Date']).dt.date

    return df
```

```python
from google.cloud import bigquery

# Replace this with your service account key file path
service_account_key_path = "StockBiqQueryKey.json"

# Initialize the BigQuery client
client = bigquery.Client.from_service_account_json(service_account_key_path)
```

```python
# Define the schema for your table
schema = [
    bigquery.SchemaField('code', 'STRING'),
    bigquery.SchemaField('name', 'STRING'),
    bigquery.SchemaField('tradingVolume', 'INTEGER'),
    bigquery.SchemaField('numberOfTrades', 'INTEGER'),
    bigquery.SchemaField('tradingAmount', 'INTEGER'),
    bigquery.SchemaField('openingPrice', 'FLOAT'),
    bigquery.SchemaField('highestPrice', 'FLOAT'),
    bigquery.SchemaField('lowestPrice', 'FLOAT'),
    bigquery.SchemaField('closingPrice', 'FLOAT'),
    bigquery.SchemaField('priceChangeDifference', 'FLOAT'),
    bigquery.SchemaField('lastRevealedBuyingPrice', 'FLOAT'),
    bigquery.SchemaField('lastRevealedBuyingQuantity', 'INTEGER'),
    bigquery.SchemaField('lastRevealedSellingPrice', 'FLOAT'),
    bigquery.SchemaField('lastRevealedSellingQuantity', 'INTEGER'),
    bigquery.SchemaField('priceToEarningsRatio', 'FLOAT'),
    bigquery.SchemaField('date', 'DATE')
]

# Define the dataset and table ID
dataset_id = 'PythonStock'  # Replace with your dataset ID
table_id = 'DailyStock_PE'      # Replace with your table ID

# Create a dataset reference
dataset_ref = client.dataset(dataset_id)

# Create the table with the specified schema
table_ref = dataset_ref.table(table_id)
table = bigquery.Table(table_ref, schema=schema)
table = client.create_table(table)  # Create the table

print(f"Table {table_id} created in dataset {dataset_id}.")
```

```python
# Define the start date and end date
start_date = datetime.date(2023, 9, 1)
end_date = datetime.date(2023, 10, 3)

# Generate a date range between start_date and end_date
date_range = pd.date_range(start=start_date, end=end_date)

# Filter the date range to keep only weekdays (Monday to Friday)
workdays = [date for date in date_range if date.weekday() < 5]

# Convert the workdays to a list of strings in the format 'YYYY-MM-DD'
workdays_str = [date.strftime('%Y%m%d') for date in workdays]
```

```python
import time
from tqdm import tqdm

# Initialize an empty DataFrame to store the concatenated data
all_data = pd.DataFrame()

# Initialize the tqdm progress bar
for date_str in tqdm(workdays_str):
    try:
        data = StockTable(date_str)
        
        # Concatenate the data to the all_data DataFrame
        all_data = pd.concat([all_data, data], ignore_index=True)
        time.sleep(1)

    except Exception as e:
        pass

print("Finish Collecting!!")

#Insert into BigQuery
job = client.load_table_from_dataframe(data, table_ref)
job.result()  #等待寫入完成

print("Mission Complete!!")
```
